Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
df.columns
```

```python
def rating(rating):
  if rating >= 8:
    return 'Excellent'
  elif rating >= 6:
    return 'Good'
  else:
    return 'Average'
```

```python
df['rat_cat'] = df['Rating'].apply(rating)
```

Find Unique Values From Genre

```python
df['Genre'].unique()
```

```python
l1 = []
for value in df['Genre']:
  l1.append(value.split(','))
```

```python
one_d = []
for item in l1:
  for value in item:
    one_d.append(value)
```

```python
one_d
```

```python
uni_list = []
for value in one_d:
  if value not in uni_list:
    uni_list.append(value)
```
